```python
# Assuming you uploaded a CSV file named 'my_data.csv'

df = pd.read_csv('chatgpt_reviews.csv')
df.head(10)
```

```python
df.info()
```

```python
def get_sentiment(rating):
    if rating in [4, 5]:
        return 'positive'
    elif rating == 3:
        return 'neutral'
    elif rating in [1, 2]:
        return 'negative'
    else:
        return None

df['sentiment'] = df['Ratings'].apply(get_sentiment)
display(df.head())
```

```python
df['sentiment'].value_counts()
```

```python
import matplotlib.pyplot as plt

sentiment_counts = df['sentiment'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%2.2f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribution of Sentiments')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
#Postive feedback is almore 90% out of 100% on chatgpt
```

```python
df['Formatted_Review_Date'] = df['Review Date'].dt.strftime('%Y-%m-%d')
display(df.head())
```

```python
# Analysis on recent 30 days review to get latest feedback on chatgpt
```

```python
from datetime import timedelta

# Ensure 'Review Date' is in datetime format
df['Review Date'] = pd.to_datetime(df['Review Date'])

# Find the latest review date
latest_review_date = df['Review Date'].max()

# Calculate the cutoff date for the last 30 days
cutoff_30_days = latest_review_date - timedelta(days=30)

# Filter the DataFrame for the last 30 days
df_recent_30_days = df[df['Review Date'] >= cutoff_30_days]

display(df_recent_30_days.head())
```

```python
sentiment_counts_30_days = df_recent_30_days['sentiment'].value_counts()
print(sentiment_counts_30_days)
```

```python
# Calculate the cutoff date for the last 60 days
cutoff_60_days = latest_review_date - timedelta(days=60)

# Filter the DataFrame for the last 60 days
df_recent_60_days = df[df['Review Date'] >= cutoff_60_days]

display(df_recent_60_days.head())
```

Here's the sentiment distribution for these recent 60 days:

```python
sentiment_counts_60_days = df_recent_60_days['sentiment'].value_counts()
print(sentiment_counts_60_days)
```

```python
# Calculate the cutoff date for the last 90 days
cutoff_90_days = latest_review_date - timedelta(days=90)

# Filter the DataFrame for the last 90 days
df_recent_90_days = df[df['Review Date'] >= cutoff_90_days]

display(df_recent_90_days.head())
```

And the sentiment distribution for the last 90 days:

```python
sentiment_counts_90_days = df_recent_90_days['sentiment'].value_counts()
print(sentiment_counts_90_days)
```
